```python
def plot_dist(tensor):
    flat_values = np.array(tensor.flatten().cpu())
    plt.hist(flat_values, bins=100, edgecolor='black')
    plt.title("Histogram of Values")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
```

```python
def plot_dist(tensor):
    flat_values = np.array(tensor.flatten().cpu())
    sorted_values = np.sort(flat_values)
    plt.figure(figsize=(10, 7))
    plt.plot(range(len(sorted_values)), sorted_values)
    plt.xlabel('Element')
    plt.ylabel('Activation Value')
```

```python
plot_dist(key_states_hiddens_states)
```

```python
plot_dist(value_states_hiddens_states)
```
